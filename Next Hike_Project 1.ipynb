##  Outliers

```python
q1 = Final_Dataset['temp'].quantile(.25)
q3 = Final_Dataset['temp'].quantile(.75)
```

### Finding Outliers:

```python
def find_outliers(Final_Dataset, weekday):
    q1 = Final_Dataset['temp'].quantile(.25)
    q3 = Final_Dataset['temp'].quantile(.75)
    iqr = q3 - q1
    min_r = q1 - 1.5*iqr
    max_r = q3 + 1.5*iqr
    outliers_indices = Final_Dataset.index[(Final_Dataset['weekday'] < min_r) | (Final_Dataset['temp'] > max_r)]
    return outliers_indices
```

```python
#To find Outliers Indexes
find_outliers(Final_Dataset,'temp')
```
